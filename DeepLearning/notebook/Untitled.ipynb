```python
from gaps_dataset import gaps
import zipfile
import os
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
from DataLoader import gen_batch_train_data
```

```python
file_name_list, _ = gen_batch_train_data('2007_train.txt')
```

```python
data = []
for file in file_name_list:
    print(file)
    im = cv2.imread(file, cv2.IMREAD_GRAYSCALE)
    im = im.astype(np.float)
    im /= 255
    data.append(im)
data = np.asarray(data)
data = np.reshape(data, (-1, 400,400,1))
```

```python
np.save('train_set_image.npy', data)
```
