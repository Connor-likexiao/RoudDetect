```python
import cv2 as cv
from matplotlib import pyplot as plt
import PreProcess
import numpy as np
```

```python
img = cv.imread('test_img/timg.jpg',cv.IMREAD_GRAYSCALE)
img = PreProcess.equalize_hist(img)
img = PreProcess.center_avg_imp(img, ksize=20)
plt.imshow(img,cmap='gray')
```

```python
import matplotlib.gridspec as gridspec
med_img = PreProcess.med_blur(img)
avg_img = cv.blur(img,ksize=(5,5))
fig, axs = plt.subplots(nrows=1, ncols=3, constrained_layout=True, figsize=(40,20))
imgs = [img, med_img, avg_img]
for ax,image,i in zip(axs.flatten(),imgs,range(3)):
    ax.imshow(image, cmap='gray')
    ax.set_title("(%d)" % i, fontsize = 32)
plt.savefig("a.jpg")
```

```python
import RegionGrow 
rg = RegionGrow.RegionGrow(med_img)
rg.img_cut()
rg.min_pos()
img = rg.region_grow()
```

```python
import PreProcess
PreProcess.hist_segmentation(med_img)
```

```python
hist = cv.calcHist([img],[0],None,[256],[0,255])
plt.plot(hist)
plt.show()
```

```python
max_val = max(hist)
```

```python
max_index = np.where(hist == max_val)
print(max_index)
```

```python
mask = hist[0:max_index[0][0]]
plt.plot(mask)
plt.show()
```

```python
min_index = np.where(mask == min(mask))
print(min_index)
```

```python
ret,new_im = cv.threshold(img,min_index[0],255,cv.THRESH_BINARY)
print(ret)
plt.subplot(1,2,1)
plt.imshow(new_im,cmap='gray')
plt.subplot(1,2,2)
plt.imshow(img,cmap='gray')
plt.show()
```

```python
k = np.ones(shape=(5,5))
dilate_img = cv.dilate(new_im, k)
k = np.ones(shape=(3,3))
erode_img = cv.erode(dilate_img,k)

fig, axs = plt.subplots(nrows=1, ncols=3, constrained_layout=True, figsize=(40,20))
imgs = [new_im, dilate_img, erode_img]
for ax,image,i in zip(axs.flatten(),imgs,range(3)):
    ax.imshow(image, cmap='gray')
    ax.set_title("(%d)" % i, fontsize = 32)
plt.savefig("a.jpg")
```
