```python
import cv2 as cv
from matplotlib import pyplot as plt
import PreProcess
import numpy as np
import RegionGrow 

img = cv.imread('test_img/timg.jpg',cv.IMREAD_GRAYSCALE)
img = PreProcess.equalize_hist(img)
img = PreProcess.center_avg_imp(img, ksize=20)
plt.imshow(img,cmap='gray')


import matplotlib.gridspec as gridspec
med_img = PreProcess.med_blur(img)
avg_img = cv.blur(img,ksize=(5,5))
fig, axs = plt.subplots(nrows=1, ncols=3, constrained_layout=True, figsize=(40,20))
imgs = [img, med_img, avg_img]
for ax,image,i in zip(axs.flatten(),imgs,range(3)):
    ax.imshow(image, cmap='gray')
    ax.set_title("(%d)" % i, fontsize = 32)
plt.savefig("a.jpg")

rg = RegionGrow.RegionGrow(med_img)
rg.img_cut()
rg.min_pos()
img = rg.region_grow()
plt.imshow(img, cmap="gray")
```

```python
plt.imshow(img,cmap='gray')
plt.show()
img = img.astype(np.uint8)
```

```python

x = cv.Sobel(img,cv.CV_16S,1,0) # 
y = cv.Sobel(img,cv.CV_16S,0,1)  
  
absX = cv.convertScaleAbs(x)   # 转回uint8  
absY = cv.convertScaleAbs(y)  
  
dst = cv.addWeighted(absX,0.5,absY,0.5,0)  

plt.figure(figsize=(20,10))
plt.imshow(dst, cmap='gray')
plt.show()
```

```python
plt.figure(figsize=(20,10))
gray_lap = cv.Laplacian(img,cv.CV_16S,ksize = 3)
dst = cv.convertScaleAbs(gray_lap)
plt.imshow(dst, cmap='gray')
plt.show()
```
